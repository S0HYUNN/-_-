```python
# csv í•©ì¹˜ê¸° -> person íŒŒì¼ ìƒì„±
# ì´ë¦„ì— . <- ìˆìœ¼ë©´ ì˜¤ë¥˜ ìƒê¸¸ ìˆ˜ ìˆì–´ ëª¨ë‘ _ ë¡œ ìˆ˜ì •

# Major
seer__eul = pd.read_csv("seer__eul.csv", low_memory=False) # __ <- 2 under bar
w_woo_n = pd.read_csv("w_woo.n.csv", low_memory=False)
hoonxxy = pd.read_csv("hoonxxy.csv", low_memory=False)
kkang_stylist = pd.read_csv("kkang.stylist.csv", low_memory=False)

Major_person = pd.concat([seer__eul,w_woo_n,hoonxxy,kkang_stylist], ignore_index = True)
save_to_csv(Major_person, "Major_person.csv")

# minor
babyjongok = pd.read_csv("babyjongok.csv", low_memory=False)
soominnnnn = pd.read_csv("soominnnnn.csv", low_memory=False)
jun_seonng = pd.read_csv("jun_seonng.csv", low_memory=False)
zzjun_k = pd.read_csv("zzjun.k.csv", low_memory=False)
h_yun_b = pd.read_csv("h.yun_b.csv", low_memory=False)
jiseong = pd.read_csv("jiseong.csv", low_memory=False)

minor_person = pd.concat([babyjongok,soominnnnn,jun_seonng,zzjun_k,h_yun_b,jiseong], ignore_index = True)
save_to_csv(minor_person, "minor_person.csv")

# total
total_person = pd.concat([Major_person, minor_person], ignore_index = True)
save_to_csv(total_person, "total_person.csv")
```

```python
# csv í•©ì¹˜ê¸° -> media íŒŒì¼ ìƒì„±

# Major
seer__eul_media = pd.read_csv("seer__eul_media.csv", low_memory=False) # __ <- 2 under bar
w_woo_n_media = pd.read_csv("w_woo.n_media.csv", low_memory=False)
hoonxxy_media = pd.read_csv("hoonxxy_media.csv", low_memory=False)
kkang_stylist_media = pd.read_csv("kkang.stylist_media.csv", low_memory=False)

Major_media = pd.concat([seer__eul_media,w_woo_n_media,hoonxxy_media,kkang_stylist_media], ignore_index = True)
save_to_csv(Major_media, "Major_media.csv")

# minor
babyjongok_media = pd.read_csv("babyjongok_media.csv", low_memory=False)
soominnnnn_media = pd.read_csv("soominnnnn_media.csv", low_memory=False)
jun_seonng_media = pd.read_csv("jun_seonng_media.csv", low_memory=False)
zzjun_k_media = pd.read_csv("zzjun.k_media.csv", low_memory=False)
h_yun_b_media = pd.read_csv("h.yun_b_media.csv", low_memory=False)
jiseong_media = pd.read_csv("jiseong_media.csv", low_memory=False)

minor_media = pd.concat([babyjongok_media,soominnnnn_media,jun_seonng_media,zzjun_k_media,h_yun_b_media,jiseong_media], ignore_index = True)
save_to_csv(minor_media, "minor_media.csv")

# total
total_media = pd.concat([Major_media, minor_media], ignore_index = True)
save_to_csv(total_media, "total_media.csv")
```
